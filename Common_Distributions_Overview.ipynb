
# Overview of Common Distributions: Normal, Binomial, and Poisson (with Python Implementation)

In this notebook, we will explore three common probability distributions:
- Normal Distribution
- Binomial Distribution
- Poisson Distribution

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm, binom, poisson, shapiro
import seaborn as sns
```


## 1. Normal Distribution 

A continuous distribution that is symmetric around the mean, shaped like a bell curve.

### Simple Example:

Heights of people, IQ scores, and blood pressure typically follow a normal distribution.


```python
# Generate Normal Distribution (Titanic 'age' data)
titanic = sns.load_dataset('titanic')
age_data = titanic['age'].dropna()

# Plot
sns.histplot(age_data, kde=True)
plt.title('Age Distribution (Normal Approximation)')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()
```


### Checking Normality: Shapiro-Wilk Test

Statistical test to check if data is normally distributed.
- Null Hypothesis (H0): Data is normal.
- Alternative Hypothesis (H1): Data is not normal.


```python
# Shapiro-Wilk test
stat, p_value = shapiro(age_data)
print(f"Shapiro-Wilk Test p-value: {p_value:.4f}")

if p_value > 0.05:
    print("Data is likely normally distributed.")
else:
    print("Data is likely NOT normally distributed.")
```


## 2. Binomial Distribution ðŸŽ¯

A discrete distribution modeling number of successes in fixed number of independent trials.

### Simple Example:

Flipping a fair coin 5 times and finding the probability of exactly 3 heads.


```python

# Binomial Distribution Example
from scipy.stats import binom

# Parameters
n = 10  # Number of customers emailed
p = 0.2  # Probability of a customer clicking the link
k = 2  # Number of clicks

# Probability Mass Function
prob = binom.pmf(k, n, p)
print(f"Probability that exactly 2 customers click the link: {prob:.2f}")
```

```python

# Binomial Distribution Plot
import numpy as np
x = np.arange(0, n+1)
y = binom.pmf(x, n, p)
plt.bar(x, y)
plt.title('Binomial Distribution: Email Clicks (n=10, p=0.2)')
plt.xlabel('Number of Customers Clicking')
plt.ylabel('Probability')
plt.show()
```


## 3. Poisson Distribution ðŸ“ž

A discrete distribution modeling the number of times an event occurs in a fixed interval.

### Formula:

\[
P(X = k) = \frac{e^{-\lambda} \lambda^k}{k!}
\]

Where:
- \( \lambda \) = average number of occurrences
- \( k \) = number of occurrences

### Simple Example:

Number of restaurant orders per minute (say, 4 orders per minute).


```python

# Poisson Distribution Example
from scipy.stats import poisson

# Parameters
lambda_ = 4  # Average orders per minute
k = 2  # Actual number of orders

# Probability
prob = poisson.pmf(k, mu=lambda_)
print(f"Probability of receiving exactly 2 orders: {prob:.2f}")
```

```python
# Poisson Distribution Plot
x = np.arange(0, 10)
y = poisson.pmf(x, mu=lambda_)
plt.bar(x, y)
plt.title('Poisson Distribution (Î»=4)')
plt.xlabel('Number of Orders')
plt.ylabel('Probability')
plt.show()
```
